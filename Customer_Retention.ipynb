```python
import pandas as pd

df = pd.read_csv('/content/online_retail_II.xlsx.csv')
df.head()
```

```python
df.drop_duplicates(inplace=True)

print("Shape after removing duplicates:", df.shape)
```

```python
df = df.dropna(subset=['Customer ID'])

print("Shape after removing missing Customer IDs:", df.shape)
```

```python
df = df[~df['Invoice'].astype(str).str.startswith('C')]

print("Shape after removing cancelled orders:", df.shape)
```

```python
df = df[df['Quantity'] > 0]

print("Shape after removing negative quantities:", df.shape)
```

```python
df = df[df['Price'] > 0]

print("Shape after removing negative prices:", df.shape)
```

```python
df['Revenue'] = df['Quantity'] * df['Price']
```

```python
df
```

```python
df.head()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.info()
```

```python
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')
```

```python
df[['InvoiceDate', 'InvoiceMonth']].head()
```

```python
cohort = df.groupby('Customer ID')['InvoiceMonth'].min()
```

```python
cohort.head()
```

```python
df['CohortMonth'] = df['Customer ID'].map(cohort)
```

```python
df[['Customer ID','InvoiceMonth','CohortMonth']].head()
```

```python
invoice_year = df['InvoiceMonth'].dt.year
invoice_month = df['InvoiceMonth'].dt.month

cohort_year = df['CohortMonth'].dt.year
cohort_month = df['CohortMonth'].dt.month
```

```python
df['CohortIndex'] = (
    (invoice_year - cohort_year) * 12 +
    (invoice_month - cohort_month) + 1
)
```

```python
df[['Customer ID','InvoiceMonth','CohortMonth','CohortIndex']].head(10)
```

```python
cohort_data = df.groupby(
    ['CohortMonth','CohortIndex']
)['Customer ID'].nunique().reset_index()
```

```python
cohort_data.head()
```

```python
cohort_table = cohort_data.pivot(
    index='CohortMonth',
    columns='CohortIndex',
    values='Customer ID'
)
```

```python
cohort_table
```

```python
retention = cohort_table.divide(cohort_table.iloc[:,0], axis=0)
```

```python
retention = retention.round(3) * 100
```

```python
retention
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

im = ax.imshow(retention, aspect='auto')

# Axis labels
ax.set_xticks(range(len(retention.columns)))
ax.set_xticklabels(retention.columns)

ax.set_yticks(range(len(retention.index)))
ax.set_yticklabels(retention.index.astype(str))

plt.xlabel("Months Since First Purchase")
plt.ylabel("Customer Cohort (First Purchase Month)")
plt.title("Customer Retention Cohort Analysis")

# Add percentage values inside each cell
for i in range(retention.shape[0]):
    for j in range(retention.shape[1]):
        value = retention.iloc[i, j]
        if not pd.isna(value):
            ax.text(
                j,
                i,
                f"{value:.1f}%",
                ha="center",
                va="center",
                fontsize=8
            )

plt.colorbar(im, label="Retention Rate (%)")

plt.tight_layout()
plt.show()
```

```python
top_country = (
    df.groupby("Country")["Revenue"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_country.plot(kind='bar', figsize=(10,5))

plt.title("Top 10 Countries by Revenue")
plt.xlabel("Country")
plt.ylabel("Revenue")

plt.show()
```

```python
top_products = (
    df.groupby("Description")["Revenue"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_products.plot(kind='bar', figsize=(12,5))

plt.title("Top 10 Products")
plt.xlabel("Product")
plt.ylabel("Revenue")

plt.xticks(rotation=45)

plt.show()
```

```python
monthly = (
    df.groupby("InvoiceMonth")["Revenue"]
      .sum()
)

monthly.plot(figsize=(12,5))

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.show()
```
